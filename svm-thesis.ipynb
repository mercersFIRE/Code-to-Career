```python
import pandas as pd

# Load the dataset
data_path="/kaggle/input/cf-data/codeforces_user_data_with_category_counts.csv"
df = pd.read_csv(data_path)
df = df.drop(columns=['User ID'])

# Show basic info
#print(df)

# Display the first few rows
#df.head()
#print(df.isnull().sum())
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Current Job Status'] = label_encoder.fit_transform(df['Current Job Status'])  # Change 'target_column' to your target column
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = df.drop(columns=['Current Job Status'])  # Features
y = df['Current Job Status']  # Target variable
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2,random_state=11)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

svm_model = SVC(kernel='linear')
svm_model.fit(X_train, y_train)

# Predictions
y_pred_svm = svm_model.predict(X_test)

# Accuracy
print(f"SVM Accuracy: {accuracy_score(y_test, y_pred_svm)}")
```
